```python
df.columns = df.columns.str.lower() ## BY THIS THE UPPER CASE WILL CHANGE INTO THE  LOWER CASE 
df.columns=df.columns.str.replace(' ','_') ##  WE WILL CALL THIS METHOD AS SNAKE CASING , IT IS EASIER TO READ AND REFER ,this we will use for the headings contains space by that we will not get errors while running in the sql or python
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'}) ## to rename  we will use THIS CMD.
```

```python
df.columns 
```

```python
# create a column age_group
labels = ['Young Adult','Adult','Middle-aged','Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels =labels)
```

```python
df[['age','age_group']].head(10)
```

```python
# create column  purchase_frequency_days
frequency_mapping = {
    'Fortnightly' : 14,
    'Weekly' : 7,
    'Monthly' : 30,
    'Quartely' : 90,
    'Bi-Weekly' : 14,
    'Annually' : 365,
    'Every 3 Months' : 90
}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)


```
